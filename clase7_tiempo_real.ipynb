En el deporte moderno, el anÃ¡lisis post-partido no es suficiente. Los entrenadores necesitan datos **AHORA**. Hoy simularemos un entorno de **Streaming** de datos.

## Objetivo

## Simulador de Partido

Vamos a crear un pequeÃ±o script que genere eventos aleatorios de un partido de fÃºtbol cada segundo (Pase, Gol, Falta).

```python
import time
import random
from IPython.display import clear_output

# Agentes posibles del juego
eventos = ['Pase', 'Pase', 'Pase', 'Tiro', 'Gol', 'Falta', 'Fuera de Lugar']
equipos = ['Local', 'Visitante']

# Estado inicial del partido
marcador = {'Local': 0, 'Visitante': 0}
posesion = {'Local': 0, 'Visitante': 0} # Contador de pases

print("Â¡Inicio del partido simulado! (DuraciÃ³n: 20 segundos)")
time.sleep(2)

# Bucle de simulaciÃ³n (Streaming Loop)
for minuto in range(1, 21):
    # 1. Generar Evento 'En Vivo'
    evento_actual = random.choice(eventos)
    equipo_actual = random.choice(equipos)
    
    # 2. Procesar Dato (Stream Processing)
    if evento_actual == 'Gol':
        marcador[equipo_actual] += 1
        mensaje = f"Â¡Â¡Â¡GOOOL del {equipo_actual}!!!"
    elif evento_actual == 'Pase':
        posesion[equipo_actual] += 1
        mensaje = f"Pase del {equipo_actual}"
    else:
        mensaje = f"{evento_actual} de {equipo_actual}"
        
    # 3. Calcular MÃ©tricas en Tiempo Real
    total_pases = posesion['Local'] + posesion['Visitante']
    if total_pases > 0:
        pct_local = (posesion['Local'] / total_pases) * 100
    else:
        pct_local = 50

    # 4. VisualizaciÃ³n (Dashboard en Vivo)
    clear_output(wait=True)
    print(f"â± Minuto: {minuto}")
    print("---------------------------------")
    print(f"âš½ MARCADOR: Local {marcador['Local']} - {marcador['Visitante']} Visitante")
    print(f"ðŸ“Š POSESIÃ“N: Local {pct_local:.1f}% - {100-pct_local:.1f}% Visitante")
    print("---------------------------------")
    print(f"ðŸ“¢ Ãšltimo evento: {mensaje}")
    
    # Simulamos espera de 1 segundo (tiempo real)
    time.sleep(1)
```

Imagina que esto procesa millones de datos por segundo. En la vida real, usarÃ­amos herramientas como **Apache Kafka** o **Spark Streaming**, pero la lÃ³gica es la misma:
